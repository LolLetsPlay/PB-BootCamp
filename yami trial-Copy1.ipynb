```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_pie_chart(file_path, title):
    # Step 1: Read data from CSV file
    data = pd.read_csv(file_path, encoding='utf-8')
    
    # Step 2: Categorize the data
    category_counts = data['value'].value_counts()
    
    # Step 3: Plot pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
    
    # Step 4: Add title
    plt.title(title)
    
    # Step 5: Show plot
    plt.show()

# Call the function with the appropriate file path and title
plot_pie_chart('C:/Users/YAMI/Desktop/Python Exp/fifa_eda.csv', 'Distribution of Categories in fifa_eda.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_pie_chart(file_path, title):
    # Step 1: Read data from CSV file
    data = pd.read_csv(file_path, encoding='utf-8')
    
    # Step 2: Define the specific categories and calculate their counts
    specific_categories = ['Germany', 'Argentina', 'England']
    category_counts = data['value'].value_counts()
    
    # Initialize the dictionary for pie chart data
    pie_data = {category: category_counts.get(category, 0) for category in specific_categories}
    
    # Calculate the count for 'Others'
    other_count = category_counts[~category_counts.index.isin(specific_categories)].sum()
    pie_data['Others'] = other_count
    
    # Step 3: Prepare data for the pie chart
    labels = pie_data.keys()
    sizes = pie_data.values()
    
    # Step 4: Plot pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    
    # Step 5: Add title
    plt.title(title)
    
    # Step 6: Show plot
    plt.show()

# Call the function with the appropriate file path and title
plot_pie_chart('C:/Users/YAMI/Desktop/Python Exp/fifa_eda.csv', 'Distribution of Categories in fifa_eda.csv')
```
